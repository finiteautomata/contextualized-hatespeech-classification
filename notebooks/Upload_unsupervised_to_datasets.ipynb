```python
from datasets import load_dataset
import glob
from tqdm.auto import tqdm

files = glob.glob("/content/drive/Shareddrives/jmperezdc/data/hatespeech-data/*.json")[:3]
all_ds = []

for f in tqdm(files):
    all_ds.append(
        load_dataset("json", data_files=[f])["train"]
    )
```

```python
from datasets import concatenate_datasets

ds = concatenate_datasets([t["train"] for t in all_ds])
```
